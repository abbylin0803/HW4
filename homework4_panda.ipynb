```python
 # Establish the bins for spending 
 spendingbin = [550, 575,600,625,650,675]
 budgetlabel = ['5', '4', '3', '2', '1']
 SchoolTable['Spending Tier'] = pd.cut(budgeteachstudent,spendingbin,labels=budgetlabel,right=False)
 SchoolTable
```

```python
SchoolTable.columns
```

```python
 # average score by spending tier 
 schooltable_mathavg = SchoolTable.groupby('Spending Tier').mean()['Average Math Score']
 schooltable_readavg = SchoolTable.groupby('Spending Tier').mean()['Average Reading Score']
 schooltable_mathpassrate = SchoolTable.groupby('Spending Tier').mean()['% Passing Math']
schooltable_readpassrate = SchoolTable.groupby('Spending Tier').mean()['% Passing Reading']
schooltable_bothpassrate = SchoolTable.groupby('Spending Tier').mean()['% Overall Passing']
```

```python
#plug those in to a new table 
TierPerformance = pd.DataFrame({"Ave Math":schooltable_mathavg,
                                "Ave Reading":schooltable_readavg, 
                                "Math Passing Rate": schooltable_mathpassrate, 
                                "Reading Passing Rate":schooltable_readpassrate, 
                                "Both Passing Rate":schooltable_bothpassrate

})
TierPerformance
```
